```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
data
```

```python
x=data.iloc[:,3:5].values
```

```python
x
```

```python
import scipy.cluster.hierarchy as sch
dendo=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Ecludian Distances')
plt.ylabel('Observation')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
yhc=ac.fit_predict(x)
yhc
```

```python
plt.scatter(x[yhc==0,0],x[yhc==0,1],c='r',s=300,label='Cluster1')
plt.scatter(x[yhc==1,0],x[yhc==1,1],c='b',s=300,label='Cluster2')
plt.scatter(x[yhc==2,0],x[yhc==2,1],c='g',s=300,label='Cluster3')
plt.scatter(x[yhc==3,0],x[yhc==3,1],c='y',s=300,label='Cluster4')
plt.scatter(x[yhc==4,0],x[yhc==4,1],c='magenta',s=300,label='Cluster5')
plt.title('Hierarchy Cluster')
plt.xlabel('Annual income')
plt.ylabel('Score')
plt.legend()
plt.show()
```
